```python
import math
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x_p, y_p = (int(i) for i in input().split())
x_c, y_c = (int(i) for i in input().split())
ang = int(input())
vet = [x_p, y_p]
vet2 = [x_p - x_c, y_p - y_c]
ang_r = math.pi*ang/180


def mudanca_origem(vetor):
    vetor[0] = vetor[0] - x_c
    vetor[1] = vetor[1] - y_c
    return vetor

def rotacao(vetor):
    aux1 = vetor[0]
    aux2 = vetor[1]
    vetor[0] = aux1*math.cos(ang_r) - aux2*math.sin(ang_r)
    vetor[1] = aux2*math.cos(ang_r) + aux1*math.sin(ang_r)
    return vetor

def desmudanca_origem(vetor):
    vetor[0] = vetor[0] + x_c
    vetor[1] = vetor[1] + y_c
    return vetor

def norma(vetor):
  s = 0
  for i in vetor:
    s += i**2
  s = s**0.5
  return s


a = desmudanca_origem(rotacao(mudanca_origem(vet)))
print(a)

xc = np.linspace(x_c - norma(vet2), x_c + norma(vet2), num=200)
y1 = (norma(vet2)**2 - (xc - x_c)**2)**0.5
y2 = -(norma(vet2)**2 - (xc - x_c)**2)**0.5


fig, esboco = plt.subplots(figsize=(15,15))
esboco.plot(x_p, y_p, marker='o', markersize=13, label='posição original', color='pink')
esboco.plot(x_c, y_c, marker='x', markersize=18, label='eixo de rotação', color='green')
esboco.plot(a[0], a[1], marker='o', markersize=13, label='posição final', color='purple')
esboco.plot(xc, y1 + y_c, color='black', zorder=0, linestyle=':', linewidth=1, alpha=0.4)
esboco.plot(xc, y2 + y_c, color='black', zorder=0, linestyle=':', linewidth=1, alpha=0.4)
plt.xlim([x_c - norma(vet2) - 1, x_c + norma(vet2) + 1])
plt.ylim([y_c - norma(vet2) - 1, y_c + norma(vet2) + 1])
plt.grid()
plt.legend(loc=1, fontsize=20)
esboco.tick_params(width=2, direction='inout', length=15, labelsize=16, grid_color='green', grid_alpha=0.3)
```
